<h1>Step 5 and Step 7</h1>

```python
def norm_params_ls(data, bins=25):
    min_val, max_val = data.min(), data.max()
    hist, _ = np.histogram(data, bins=bins, density=True)
    xdata = np.linspace(min_val, max_val, num=bins)
    
    par, _ = scipy.optimize.curve_fit(norm.pdf, xdata, hist,\
                                    p0=list(norm.fit(data)),\
                                    maxfev=1000)
    
    return par


def t_params_ls(data, bins=25):
    min_val, max_val = data.min(), data.max()
    hist, _ = np.histogram(data, bins=bins, density=True)
    xdata = np.linspace(min_val, max_val, num=bins)
    
    par, _ = scipy.optimize.curve_fit(t.pdf, xdata, hist,\
                                    p0=list(t.fit(data)),\
                                    maxfev=1000)
    
    return par


def weib_params_ls(data, bins=25):
    min_val, max_val = data.min(), data.max()
    hist, _ = np.histogram(data, bins=bins, density=True)
    xdata = np.linspace(min_val, max_val, num=bins)
    
    par, _ = scipy.optimize.curve_fit(exponweib.pdf, xdata, hist,\
                                    p0=list(exponweib.fit(data, f0=10)),\
                                    maxfev=1000)
    
    return par
```

```python
plt.figure(figsize=(10, 8))
plt.title("Sulfate")
sns.histplot(sulfate, kde=False, bins=25, stat='density')
x = np.linspace(np.min(sulfate), np.max(sulfate))

plt.plot(x, norm.pdf(x, *norm.fit(sulfate)), 'r--', label="Normal MM")

plt.plot(x, norm.pdf(x, *norm_params_ls(sulfate)), 'y--', label="Normal LS")

plt.plot(x, exponweib.pdf(x, *exponweib.fit(sulfate, f0=15)),'b--', label='Exponentiated Weibull MM')

plt.plot(x, exponweib.pdf(x, *weib_params_ls(sulfate)), 'g--', label='Exponentiated Weibull LS')
        
plt.plot(x, t.pdf(x, *t.fit(sulfate)), 'k--', label='t-student MM')

plt.plot(x, t.pdf(x, *t_params_ls(sulfate)), 'c--', label='t-student LS')

plt.legend()
plt.savefig("Sulfate_MM.png", format='PNG')
```

Based on the test results, it's hard to choose the exact distribution, but according to the pvalue of CramerVonMises test of exponentiated Weibull method I chose this distribution with least squares estimation parameters

<h1>Step 6</h1>

```python
# Calculation of quantiles
percs = np.linspace(5, 95, 19)
qn_first = np.percentile(sulfate, percs)
qn_norm = scipy.stats.norm.ppf(percs / 100.0, *norm_params_ls(sulfate))


min_qn = np.min([qn_first.min(), qn_norm.min()])
max_qn = np.max([qn_first.max(), qn_norm.max()])
x = np.linspace(min_qn - 50, max_qn + 50)

# Building a quantile biplot
plt.figure(figsize=(10, 8))
plt.plot(qn_first, qn_norm, ls="", marker="o", markersize=8)
plt.plot(x, x, color="k", ls="--")
plt.xlabel('Empirical distribution')
plt.ylabel('Theoretical distribution')

plt.title('QQ-plot Sulfate - normal LS')
plt.savefig("Sulfate_QQ.png")
```

```python
# Calculation of quantiles
percs = np.linspace(5, 95, 19)
qn_first = np.percentile(organic_carbon, percs)
qn_exp_weib = scipy.stats.exponweib.ppf(percs / 100.0, *exponweib.fit(organic_carbon, f0=15))

min_qn = np.min([qn_first.min(), qn_exp_weib.min()])
max_qn = np.max([qn_first.max(), qn_exp_weib.max()])
x = np.linspace(min_qn - 3, max_qn + 3)

# Building a quantile biplot
plt.figure(figsize=(10, 8))
plt.plot(qn_first, qn_exp_weib, ls="", marker="o", markersize=8)
plt.plot(x, x, color="k", ls="--")
plt.xlabel('Empirical distribution')
plt.ylabel('Theoretical distribution')

plt.title('QQ-plot Organic carbon - exponentiated weibull MM')
plt.savefig("Organic_carbon_QQ.png")
```
